```python
import numpy as np
import time
```

```python
#execution time
size = 10_000_000

#python list
py_list = list(range(size))
start = time.time()
sq_list = [x**2 for x in py_list]
end = time.time()
print(f"python list time = {end-start} seconds")

#numpy arrays
np_arr = np.array(py_list)
newstart = time.time()
# vectorisation to do calc on a numpy array
sq_array = np_arr ** 2  # easy way to calc the square
newend = time.time()
print(f"Numpy array time = {newend - newstart} seconds")
```
