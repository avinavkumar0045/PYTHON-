```python
import requests

URL ="https://www.scrapethissite.com/pages/simple/"
res = requests.get(URL)
```

```python
if(res.status_code == 200):
    #print(res.text)
    #print(res.content)
    print(res.headers)
with open("scraped_data/data1.html" , "w") as f: #opening a file and storing the data i got in response 
    f.write(res.text)
```

```python
#BeautifulSoup
from bs4 import BeautifulSoup

with open("scraped_data/data1.html","r") as f:
    html_content= f.read()

soup = BeautifulSoup(html_content, 'lxml') #we have created a opbject of BeautifulSoup class
```

```python
#to find all the countries , we have noticed they are stored int he h3 tag
all_h3  = soup.find_all("h3") #storing the countries name

all_countries =  [] #to store all the data which we will extract later

for h3 in all_h3:
    name = h3.get_text(strip = True) #stroing the text from the tag after removing all the spaces 
   # print(h3.find_parent("div"))
    population= (h3.find_next("div").select("span.country-population")[0].get_text(strip = True)) #gives the country info as in the original website

    #print(h3.attrs)

    all_countries.append([name , population])
```

```python
import pandas as pd 
df = pd.DataFrame(all_countries, columns=['Name','Population'])
```

```python
df.to_csv("Cleaned_data/data.csv", index=False) #finally we stotr the clean data
```
