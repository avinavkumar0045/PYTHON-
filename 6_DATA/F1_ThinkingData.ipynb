```python
import json
```

```python
#load the data
def load_data(filename):
    with open(filename , "r") as f:
        data = json.load(f)
    return data
```

```python
data = load_data("F1_text.json")
print(data)
print(type(data))
```

```python
#Clean & structure the data
def clean_data(data):
    text_to_num = {"one":1 ,"two":2 ,"three":3 , "four":4 ,"five":5}
    cleaned_data = []
    unique_users = set()
    
    for user in data:
        #clean rating-data consistency
        raw_rating = str(user["rating"]).strip().lower() #with strip remove white spaces 
        if( raw_rating in text_to_num):
            raw_rating = text_to_num[raw_rating]
        user["rating"] = raw_rating

        #Handle missing values
        raw_age = user.get("age")
        if(raw_age == None):
            user["age"] = None

        #Deduplication
        if(user["name"].strip() in unique_users):
            continue

        unique_users.add(user["name"])
        cleaned_data.append(user)


    return cleaned_data
                
        
        
```

```python
data = clean_data(data)
```

```python
#Get meaning full insights from the data
def get_insights(data):

    #avg rating
    tot_rating = 0
    for user in data:
       tot_rating += float(user["rating"])
    print(f"The avrage rating of the users is : {tot_rating / len(data)}")

    #percentage of users with the less rating
    poor_rating  = 0
    for user in data:
        if( float(user["rating"])  < 3.0 ) :
            poor_rating += 1;
    print(f"Percentage of user with lesser rating is : {poor_rating / len(data)* 100}% ")
    
    
```

```python
get_insights(data)
```

```python
def get_recommendation(data):
    recommendation = []

    for user in data:
        curr_recomm = {}
        curr_recomm["name"] = user["name"]

        if( float(user["rating"]) >= 4):
            curr_recomm["brand"] = "Apple"
        else:
            curr_recomm["brand"] = "Samsung"
            
        recommendation.append(curr_recomm)
        
    return recommendation
```

```python
get_recommendation(data)
```
