```python
#Filtering of data

df [(df["aqi"] > 100)  & (df["temperature"] >37)]

#chaining 
df[df["aqi"] > 200] [["city" ,"aqi"]]
```

```python
#query method

df.query("aqi > 100 & temperature > 30") #within query write as much as you want to write 
aqi_val = 100
df.query("aqi > @aqi_val & temperature > 30") 
```

Cleaning data 


```python
#cleaning data - handle missing data
df = pd.read_csv("raw_data.csv")
df.isnull() # to detect where is null values

df.isnull().sum() #return the number of missing values 

df.dropna()
df.dropna(axis = 1)

df.fillna(0)

age_mean= df["age"].mean()

clean_data = df.copy() #dont change the original data
clean_data["age"].fillna(age_mean)
df.ffill() #fills with the previous data
df.bfill() #fills with the upcomming data


```

```python
#Handle duplicates

df[["country","gender"]].duplicated()
df2 = df.copy()

df.drop_duplicates(inplace = True)
```

```python
#Data types

df.dtypes

df2 = df.copy()
df2 = df2.fillna(0) #first fill the empty 
df2["age"] = df2["age"].astype("int64").copy()
df2.dtypes
```
